```python
# Dependencies and Setup
import pandas as pd

# File to Load 
pymoli = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(pymoli)
purchase_df.head()
```

## Player Count

* Display the total number of players


```python
#find the total number of player
Total_players = len(purchase_df["SN"].unique())
# create a data frame with total_players count 
player_count = pd.DataFrame({"Total players":Total_players},index = [0])
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
number_items = purchase_df["Item ID"].nunique()
avg_price = purchase_df["Price"].mean()
total_purchases = purchase_df["Purchase ID"].nunique()
total_revenue = purchase_df.Price.sum()

# Create df
purchasing_analysis = pd.DataFrame({"Number of Unique Items": [number_items],
                              "Average Price": [avg_price],
                              "Number of Purchases": [total_purchases],
                              "Total Revenue": [total_revenue]})

# Format
pd.options.display.float_format = '${:,.2f}'.format

# Display df
purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
groupby_sn_gender = purchase_df.groupby(["SN","Gender"])
genderdemo_df = pd.DataFrame(groupby_sn_gender.size())

pymoli_by_gender = pd.DataFrame(genderdemo_df.groupby(["Gender"]).count())
pymoli_by_gender.columns= ["Total Count"]

pymoli_by_gender["Percentage of Players"] = round(100*(pymoli_by_gender["Total Count"]/pymoli_by_gender["Total Count"].sum()),2)

pymoli_by_gender = pymoli_by_gender[["Percentage of Players","Total Count"]]


female_count = pymoli_by_gender["Total Count"].loc["Female"]

male_count = pymoli_by_gender["Total Count"].loc["Male"]

other_count = pymoli_by_gender["Total Count"].loc["Other / Non-Disclosed"]

pymoli_by_gender
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Calculations
gender_counts = purchase_df.groupby(["Gender"]).count()["Price"].rename("Purchase Count")
gender_avg = purchase_df.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")
gender_total_purchases = purchase_df.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")
cleaned = gender_total_purchases / pymoli_by_gender["Total Count"]

# Create df
gender_data = pd.DataFrame({"Purchase Count": gender_counts, "Average Purchase Price": gender_avg, "Total Purchase Value": gender_total_purchases, "Normalized Totals": cleaned})

# Format
gender_data["Average Purchase Price"] = gender_data["Average Purchase Price"].map("${:,.2f}".format)
gender_data["Total Purchase Value"] = gender_data["Total Purchase Value"].map("${:,.2f}".format)
gender_data ["Purchase Count"] = gender_data["Purchase Count"].map("{:,}".format)
gender_data["Avg Total Purchase per Person"] = gender_data["Normalized Totals"].map("${:,.2f}".format)
gender_data = gender_data.loc[:, ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]

# Display df
gender_data
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# slice the data using pd.cut and  Categorize the existing players based on age_bins  
purchase_df["Age Group"] = pd.cut(purchase_df["Age"],bins, labels=group_names)
purchase_df

# do a groupby on Age Group 
groupby_age = purchase_df.groupby("Age Group")
#calculate total count and percenatge of players by age category
totalplayers_age = groupby_age["SN"].nunique()
Percent_of_players = (totalplayers_age/Total_players)*100
# create a DataFrame to hold the above results 
pymoli_byage = pd.DataFrame({"Percentage of players":Percent_of_players,"Total count":totalplayers_age})
# give the displayed data cleaner formatting
pymoli_byage.index.name = None
pymoli_byage["Percentage of players"] = pymoli_byage["Percentage of players"].map("{:.2f}".format)
pymoli_byage
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# do calculations to obtain purchase count, avg. purchase price, total purchase value , average purchase per person
purchase_count = groupby_age["Age"].count()
avg_purchase = groupby_age["Price"].mean()
total_purchase_value = groupby_age["Price"].sum()
avg_purchase_person =(groupby_age["Price"].sum()/totalplayers_age)

# create a DataFrame to hold the above results 
Purchase_analysis_age = pd.DataFrame({"Purchase count":purchase_count,"Average Purchase Price":avg_purchase,
                           "Total Purchase value":total_purchase_value,"Average Purchase Total per person":avg_purchase_person})
# give the displayed data cleaner formatting 
Purchase_analysis_age["Average Purchase Price"] = Purchase_analysis_age["Average Purchase Price"].map("${:.2f}".format)
Purchase_analysis_age["Total Purchase value"]  = Purchase_analysis_age["Total Purchase value"].map("${:,.2f}".format)
Purchase_analysis_age["Average Purchase Total per person"] = Purchase_analysis_age["Average Purchase Total per person"].map("${:.2f}".format)
Purchase_analysis_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

groupby_sn = purchase_df.groupby("SN")

# do calculations to obtain purchase count, avg. purchase price, total purchase value 
purchase_count = groupby_sn["Age"].count()
avg_purchase = groupby_sn["Price"].mean()
total_purchase_value = groupby_sn["Price"].sum()

# create a DataFrame to hold the results 
Top_spenders = pd.DataFrame({"Purchase count":purchase_count,"Average Purchase Price":avg_purchase,
                           "Total Purchase value":total_purchase_value})

# Sort the total purchase value column in descending order 
Top_spenders = Top_spenders.sort_values("Total Purchase value", ascending = False)

# do a clean formatting 
Top_spenders["Average Purchase Price"] = Top_spenders["Average Purchase Price"].map("${:.2f}".format)
Top_spenders["Total Purchase value"]  = Top_spenders["Total Purchase value"].map("${:,.2f}".format)

# dDisplay a preview of the summary data frame
Top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve the Item ID, Item Name, and Item Price columns   
mostpopular_df = purchase_df.loc[:,["Item ID","Item Name","Price"]]

# Group by Item ID and Item Name. 
groupby_popular = purchase_df.groupby(["Item ID","Item Name"])

# Perform calculations to obtain purchase count, item price, and total purchase value
purchase_count = groupby_popular["Age"].count()
avg_purchase = groupby_popular["Price"].mean()
total_purchase_value = groupby_popular["Price"].sum()

# create a DataFrame to hold the results 
mostpopular_items = pd.DataFrame({"Purchase count":purchase_count,"Average Purchase Price":avg_purchase,
                                   "Total Purchase value":total_purchase_value})

# Sort the purchase count  column in descending order 
mostpopular_items = mostpopular_items.sort_values("Purchase count", ascending = False)

# do a clean formatting 
mostpopular_items["Total Purchase value"]  = mostpopular_items["Total Purchase value"].map("${:,.2f}".format)
mostpopular_items["Average Purchase Price"] = mostpopular_items["Average Purchase Price"].map("${:.2f}".format)


# display the Most Popular Items summary table 
mostpopular_items.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Sort the above table by total purchase value in descending order

mostprofitable_df = mostpopular_items.sort_values(by = ["Total Purchase value"],ascending = False)

mostprofitable_df.head()
```
